```python
# ---- STEP 4: DATA CLEANING ---- #

import pandas as pd

df_clean = df.copy()  # keep original safe

# 1. Fix Likert scale responses (replace >5 with NaN)
likert_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns
df_clean[likert_cols] = df_clean[likert_cols].applymap(lambda x: x if 1 <= x <= 5 else None)

# 2. Rename long columns to short names
rename_map = {
    "Please indicate your4ment or dis4ment with the following statements about your work area/unit. Think about your hospital work area/unit.People support one another in this unit": "team_support",
    "We have enough staff to handle the workload": "staff_enough",
    "When a lot of work needs to be done quickly, we work together as ateam to get the work done": "work_together_fast",
    "Staff feel like their mistakes are held against them": "blame_culture",
    "Hospital units work well together to provide the best care for patients": "units_work_together",
    "Shift changes are problematic for patients in this hospital": "shift_problems",
    "Respondents Hospital": "hospital",
    "What is your primary work area or unit in this hospital? Select ONE answer.": "unit",
    "What is your staff position in this hospital? Select ONE answer that best describes your staff position.": "staff_position",
    "Others": "others_text"           # rename to easier name
}

df_clean.rename(columns=rename_map, inplace=True)

# 3. Handle Work Area Answers (use text from others when main is 'other')
# Normalize both fields
df_clean["unit"] = df_clean["unit"].astype(str).str.strip().str.lower()
df_clean["others_text"] = df_clean["others_text"].astype(str).str.strip().str.lower()

# Replace unit="other" with the free text if available
df_clean.loc[df_clean["unit"] == "other", "unit"] = df_clean.loc[df_clean["unit"] == "other", "others_text"]

# 4. Drop helper column now that merging is done
df_clean.drop(columns=["others_text"], inplace=True)

print("Data Cleaning Completed âœ”")

# Preview cleaned data
df_clean.head()
```

```python
# ================== STEP 5 â€” DATA OVERVIEW ==================

from IPython.display import Markdown, display

display(Markdown("## ðŸ“Š Step 5 â€” Data Overview & Missing Analysis"))

print("### Dataset Structure ###")
df_clean.info()

# Missing values count
missing_count = df_clean.isnull().sum().sort_values(ascending=False)
display(Markdown("### ðŸ” Missing Values per Column"))
display(missing_count)

# Missing values %
missing_pct = (df_clean.isnull().mean()*100).sort_values(ascending=False)
display(Markdown("### ðŸ“‰ Percentage Missing"))
display(missing_pct)


# ================== STEP 6 â€” STATISTICAL SUMMARY ==================
display(Markdown("## ðŸ“ˆ Step 6 â€” Likert Statistical Summary"))

likert_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns

likert_summary = df_clean[likert_cols].describe()
display(likert_summary)


# ================== STEP 7 â€” MEAN SCORES ==================
display(Markdown("## ðŸ§  Step 7 â€” Likert Mean Score Ranking"))

mean_scores = df_clean[likert_cols].mean().sort_values(ascending=False)
display(mean_scores)


# ================== STEP 8 â€” % AGREEMENT (3 or Above) ==================
display(Markdown("## ðŸ‘ Step 8 â€” % Positive Agreement (Score â‰¥ 3)"))

agreement_pct = df_clean[likert_cols].apply(lambda x: (x>=3).mean()*100).sort_values(ascending=False)
display(agreement_pct)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ------------------- FUNCTION TO ADD NUMBERS -------------------
def add_value_labels(ax):
    for p in ax.patches:
        value = int(p.get_height())
        ax.text(
            p.get_x() + p.get_width()/2,
            value + 0.1,
            str(value),
            ha='center',
            va='bottom',
            fontsize=9
        )


# ------------------- HOSPITAL DISTRIBUTION -------------------
display(Markdown("## ðŸ¥ Hospital Distribution"))

plt.figure(figsize=(12,6))
ax = df_clean["hospital"].value_counts().plot(kind="bar", width=0.7)
plt.title("Respondent Count by Hospital")
plt.xlabel("Hospital")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=60, ha='right')

add_value_labels(ax)

plt.text(
    0.98, 0.95,
    "KEY:\nHigher bars = better data confidence\nShows representativeness across hospitals",
    transform=ax.transAxes,
    fontsize=10,
    va='top',
    ha='right',
    bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.45)
)

plt.tight_layout()
plt.show()


# ------------------- STAFF POSITION DISTRIBUTION -------------------
display(Markdown("## ðŸ‘· Staff Position Distribution"))

plt.figure(figsize=(12,6))
ax = df_clean["staff_position"].value_counts().plot(kind="bar", width=0.7)
plt.title("Respondent Count by Staff Position")
plt.xlabel("Staff Role")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=60, ha='right')

add_value_labels(ax)

plt.text(
    0.98, 0.95,
    "KEY:\nShows which roles drive safety perception data",
    transform=ax.transAxes,
    fontsize=10,
    va='top',
    ha='right',
    bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.45)
)

plt.tight_layout()
plt.show()


# ------------------- PRIMARY WORK AREA DISTRIBUTION -------------------
display(Markdown("## ðŸ§ª Primary Work Area Distribution"))

plt.figure(figsize=(14,6))
ax = df_clean["unit"].value_counts()[:15].plot(kind="bar", width=0.7)
plt.title("Primary Work Area / Unit Distribution (Top 15)")
plt.xlabel("Unit")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=60, ha='right')

add_value_labels(ax)

plt.text(
    0.98, 0.95,
    "KEY:\nShows which units contribute the most insight\nLower units need targeted engagement",
    transform=ax.transAxes,
    fontsize=10,
    va='top',
    ha='right',
    bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.45)
)

plt.tight_layout()
plt.show()
```

```python
# STEP 9 â€” Frequency + Percentage Table For Each Likert Question

from IPython.display import Markdown

def likert_frequency_table(column):
    freq = df_clean[column].value_counts().sort_index()
    pct = df_clean[column].value_counts(normalize=True).sort_index() * 100
    return pd.DataFrame({"Frequency": freq, "Percentage": pct.round(2)})

likert_tables = {}
for col in likert_cols:
    likert_tables[col] = likert_frequency_table(col)
    display(Markdown(f"### ðŸ“Š Frequency & % for: **{col}**"))
    display(likert_tables[col])
```

```python
# STEP 10 â€” Favorability Scoring

def interpret_favorability(score):
    if score >= 4.0: return "Strongly Favorable"
    if score >= 3.5: return "Moderately Favorable"
    if score >= 3.0: return "Neutral"
    if score >= 2.5: return "Moderately Unfavorable"
    return "Strongly Unfavorable"

favorability_df = pd.DataFrame({
    "Mean Score": df_clean[likert_cols].mean(),
    "Interpretation": df_clean[likert_cols].mean().apply(interpret_favorability)
}).sort_values("Mean Score", ascending=False)

display(Markdown("## ðŸ§  Favorability Interpretation Table"))
display(favorability_df)
```

```python
# STEP 11 â€” Top 10 Positive & Negative Insights

top_positive = favorability_df.head(10)
top_negative = favorability_df.tail(10)

display(Markdown("### ðŸ’š Top Positive Safety Culture"))
display(top_positive)

display(Markdown("### ðŸ”´ Top Areas Needing Improvement"))
display(top_negative)
```

```python
# STEP 12 â€” Auto-Generate Insights

for col in likert_cols:
    mean = df_clean[col].mean()
    interpretation = interpret_favorability(mean)
    agree_pct = (df_clean[col] >= 4).mean()*100  # people who agree/strongly agree
    neutral_pct = (df_clean[col] == 3).mean()*100
    disagree_pct = (df_clean[col] <= 2).mean()*100
    
    print(f"\nðŸ¥‡ INSIGHT â€” {col}")
    print(f"Mean Score: {mean:.2f} ({interpretation})")
    print(f"Agreement: {agree_pct:.1f}% | Neutral: {neutral_pct:.1f}% | Disagreement: {disagree_pct:.1f}%")
```
